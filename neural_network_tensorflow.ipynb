```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
```

```python
df = pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
df = df.drop(["ID", "ZIP Code"], axis=1)
df = pd.get_dummies(df, columns=["Family", "Education", "Securities Account", "CD Account", "Online", "CreditCard"])
```

```python
X = df.drop(["Personal Loan"], axis=1).values
Y = df["Personal Loan"].values.reshape(-1, 1)
```

```python
X = (X - X.mean()) / X.std()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
input_dim = X.shape[1]
hidden_dim = 10
output_dim = 1

model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_dim, activation='relu', input_dim=input_dim),
    tf.keras.layers.Dense(output_dim, activation='sigmoid')
])

learning_rate = 0.01
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])

num_iterations = 1000
model.fit(X_train, Y_train, epochs=num_iterations, verbose=1)

Y_pred = model.predict(X_test)
Y_pred_binary = np.round(Y_pred)
acc = accuracy_score(Y_test, Y_pred_binary)
print(f"Accuracy: {acc}")
```
